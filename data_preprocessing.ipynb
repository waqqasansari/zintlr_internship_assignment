```python
import re

target_string = asst_s
# split
assets_list = re.findall(r"(?:[\£\$\€]{1}[,\d]+.?\d* [B-M])", target_string)
print(assets_list)
print(len(assets_list))
```

```python
markt_s = df['Market_Value'].squeeze()
target_string = markt_s
# split
mv_list = re.findall(r"(?:[\£\$\€]{1}[,\d]+.?\d* [B-M])", target_string)
print(mv_list)
len(mv_list)
```

```python
def convertList(data):
  string = data.squeeze()
  l = string.split(",")
  return l
```

```python
print(convertList(df['Name']))
```

```python
rank = convertList(df['Rank'])
name = convertList(df['Name'])
country = convertList(df['Country'])
sales = convertList(df['Sales'])
profit = convertList(df['Profit'])
assets = convertList(df['Assets'])
market_value = convertList(df['Market_Value'])
link = convertList(df['link'])
```

```python
print(f'length of name list = {len(name)}') 
print(f'length of rank list = {len(rank)}')
print(f'length of country list = {len(country)}')
print(f'length of sales list = {len(sales)}')
print(f'length of profit list = {len(profit)}')
print(f'length of assets list = {len(assets_list)}') #sus
print(f'length of market_value list = {len(mv_list)}') 
print(f'length of link list = {len(link)}')
```

```python
forbes_2000 = pd.DataFrame.from_dict(
    {'Rank': rank,
     'Name': name,
     'Country': country,
     'Sales': sales,
     'Profit': profit,
     'Assets': assets_list,
     'Market_Value': market_value,
     'Link': link
    }, orient='index')

forbes_2000 = forbes_2000.transpose()
forbes_2000
```
